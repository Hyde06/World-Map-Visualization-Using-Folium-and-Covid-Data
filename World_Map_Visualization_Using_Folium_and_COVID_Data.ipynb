# Import Libraries

```python
import json
import folium
import requests
import mimetypes
import http.client
import pandas as pd
from folium.plugins import HeatMap
from pandas.io.json import json_normalize
```

# Acess Data via API

```python
import http.client

conn = http.client.HTTPSConnection("api.covid19api.com")
payload = ''
headers = {}
conn.request("GET", "/summary", payload, headers)
res = conn.getresponse()
data = res.read()
```

Pre Processing

```python
covid1 = json.loads(data)
```

```python
pd.json_normalize(covid1['Countries'],sep=',')
```

```python
df = pd.DataFrame(covid1['Countries'])
df
```

```python
covid2 = df.drop(columns =['ID','CountryCode','Slug','Date','Premium','NewRecovered','TotalRecovered'],axis=1)
covid2
```

Generating base map

```python
m = folium.Map(tiles="Stamen Terrain",min_zoom=1.5)
m
```

Generate Choropleth Map Layer

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country_shapes=f'{url}/world-countries.json'
geo_json_data = json.loads(requests.get(country_shapes).text)
```

```python
folium.Choropleth(
    geo_data=geo_json_data,
    min_zoom =1.5,
    name = 'Covid-19',
    data = covid2,
    columns= ['Country','TotalConfirmed'],
    key_on = 'feature.properties.name',
    fill_color = 'OrRd',
    nan_fill_color = 'black',
    legend_name ='Total Confirmed COVID Cases',
).add_to(m)
m
```

Generating Circular marks

```python
covid2.update(covid2['TotalConfirmed'].map('Total Confirmed:{}'.format))
```

```python
co_ordinates = pd.read_csv('/content/drive/MyDrive/Python Project/Coordinates.csv')
co_ordinates
```

```python
covid_final = pd.merge(covid2,co_ordinates,on ='Country')
```

```python
covid_final
```

```python
def plotDot(point):
    folium.CircleMarker(location=[point.latitude,point.longitude],
                       radius=5,
                       weight=2,
                       popup=[point.Country,point.TotalConfirmed],
                       fill_color='#000000').add_to(m)
```

```python
covid_final.apply(plotDot,axis=1)
m.fit_bounds(m.get_bounds())
m
```

Generate a Heat map

```python
m1= folium.Map(tiles='StamenToner',min_zoom=2)
m1
```

```python
deaths = covid_final['TotalDeaths'].astype(float)
lat = covid_final['latitude'].astype(float)
lon = covid_final['longitude'].astype(float)
m1.add_child(HeatMap(zip(lat,lon,deaths),radius=0))
```

```python
def plotDDot(point):
    folium.CircleMarker(location=[point.latitude,point.longitude],
                       radius=5,
                       weight=2,
                       popup=[point.Country,point.TotalDeaths],
                       fill_color='#000000').add_to(m1)
```

```python
covid_final.apply(plotDDot,axis=1)
m1.fit_bounds(m1.get_bounds())
m1
```
